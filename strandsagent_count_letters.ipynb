```python
from strands import Agent, tool
from strands.models.openai import OpenAIModel

from google.colab import userdata

import logging

# configure log output
logging.getLogger("strands").setLevel(logging.INFO)
logging.getLogger("strands.models").setLevel(logging.DEBUG)
logging.getLogger("strands.tools.executors").setLevel(logging.DEBUG)
```

```python
base_url = "https://chat-ai.academiccloud.de/v1"

model_id = "qwen3-coder-30b-a3b-instruct"
```

```python
#@title API Key Configuration { display-mode: "form" }
# choose one option to set your API key

# 1. Set it directly here and hide the cell content once executed to keep your key private
api_key = "123"

# 2. Read it from an ENV
# api_key = os.environ.get("GWDG_API_KEY")
```

```python
@tool
def count_r_in_word(word: str) -> int:
  """Counts the number the letter 'r' occurs in the word and returns it as an interger."""
  return len([True for ch in word if ch.lower() == "r"])
```

```python
@tool
def count_letter_in_word(word: str, letter: str) -> int:
  """Counts the number the letter occurs in the word and returns it as an interger."""
  return len([True for ch in word if ch.lower() == letter])
```

```python
model = OpenAIModel(client_args={"api_key": api_key, "base_url": base_url}, model_id=model_id, params={"temperature": 0.0})
```

```python
prompt = """How many r and how many s are in the word StrawbeRRRiesSS?"""
```

## Agent with out tools

```python
agent_wo_tools = Agent(system_prompt="You are a helpful agent", model=model, tools=None)
```

```python
response_wo_tools = agent_wo_tools(prompt=prompt)
```

```python
print(response_wo_tools.message['content'][0]['text'])
```

## Agent with letter counting tools

### First, only use tool to count "R"s

```python
agent_w_tools = Agent(system_prompt="You are a helpful agent", model=model, tools=[count_r_in_word])
```

```python
response_w_tools = agent_w_tools(prompt=prompt)
```

```python
print(response_w_tools.message['content'][0]['text'])
```

### Now, use the more general count letters function

```python
agent_w_tools = Agent(system_prompt="You are a helpful agent", model=model, tools=[count_letter_in_word])
```

```python
response_w_tools = agent_w_tools(prompt=prompt)
```

```python
print(response_w_tools.message['content'][0]['text'])
```
